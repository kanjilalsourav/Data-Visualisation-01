```python
4.4. Outliers Treatment
For the analysis of the outlier data, we have considered the rules of cricket. The outliers exist when there are any extra runs awarded like wide or no ball. The problem is mostly related to the distribution of the run between the run scored by the batsman and the extra runs.\ as we will see. 

As we recall we had 33 outliers which had 8 and 10 runs as total_runs.
```

```python
dlv[dlv.total_runs == 8]
```

As we can see the outlier is coming because data of leg bye or bye has been added to the indvidual score of the batsman as well which is wrong because a bats man cant score if a bye or leg bye of four runs is there. There is also a a scenario where in a no ball a batsman hits 7 runs which though is theoritacally possible but quite impractical. Hence we have limited the increment of the batsman's score to 6 and added 1 extra run for the no ball.

```python
dlv[dlv.total_runs == 10]
```

Here also the outliers are arising because the extras from the wide ball has been added to the batsman's individal score. So we will change the batsman's individual score to 0 and and just keep the extra runs.

```python
def clean_total_runs(df) :
    if(df.total_runs > 7 & df.wide_runs != 0) :
        return 0, df.extra_runs
    elif(df.total_runs > 7 & df.bye_runs != 0) :
        return 0, df.extra_runs
    elif(df.total_runs > 7 & df.legbye_runs != 0) :
        return 0, df.extra_runs
    elif(df.total_runs > 7 & df.noball_runs != 0) :
        return 6, (6 + df.extra_runs)
    else :
        return df.batsman_runs, df.total_runs
```

```python
dlv['batsman_runs'], dlv['total_runs'] = zip(*dlv.apply(clean_total_runs, axis=1))
```

__Top 5 Batsmen__

```python
batsmen = dlv_new.groupby(['batsman', 'match_id'], as_index=False).sum()[['batsman','match_id','batsman_runs']]
batsmen_most_runs = batsmen.groupby('batsman', as_index=False).sum()[['batsman', 'batsman_runs']]
batsmen_5_most_runs = batsmen_most_runs.nlargest(5, 'batsman_runs')
batsmen_5_most_runs
```

```python
def calculate_batsman_details(df) :
    batsmen_list = []
    for row in df.itertuples() :
        matches = batsmen[batsmen.batsman == row.batsman]['batsman_runs'].count()
        highest = batsmen[batsmen.batsman == row.batsman]['batsman_runs'].max()
        batsmen_list.append({'name' : row.batsman, 'runs' : row.batsman_runs, 'matches' : matches, 'avg' : float(row.batsman_runs/matches), 'highest' : highest})
        
    return batsmen_list
```

```python
batsmen_5most_runs_details =  pd.DataFrame(calculate_batsman_details(batsmen_5_most_runs))
batsmen_5most_runs_details
```

```python
source = ColumnDataSource(batsmen_5most_runs_details)

plot1 = figure(x_range= batsmen_5most_runs_details.name, plot_height=450, y_axis_label="Runs", title="Batsman with most runs", toolbar_location=None)

plot1.vbar(x='name', top='runs', width=0.6, source = source, legend = 'name',
           fill_color=factor_cmap('name', palette=Spectral5, factors=batsmen_5most_runs_details.name))

plot1.xgrid.grid_line_color = None
plot1.y_range.start = 0
plot1.y_range.end = 6500
hover = HoverTool(tooltips = [('Matches','@matches'),('Runs','@runs'), ('Average', '@avg'),('Highest','@highest')])
plot1.add_tools(hover)
plot1.legend.orientation = "horizontal"
plot1.legend.location = "top_center"

show(plot1)
```

__Top 5 Bowlers__

```python
dlv_new['dismissal_kind'].unique()
```

As we know out of all the ways being dismissed, a bowler does not gets credit if the batsman gets dismissed because of being retired hurt or run out or obstructing the field. There are also nome Nan values.  Our goal is to remove thse data.

```python
dismissal_kind= ['caught', 'bowled', 'lbw', 'caught and bowled','stumped', 'hit wicket']
bowlers= dlv_new[(dlv_new['player_dismissed'].notnull()) & (dlv_new['dismissal_kind'].isin(dismissal_kind))].groupby(['bowler', 'match_id'])['bowler', 'match_id'].agg(
    {"wickets": len}).reset_index()
bowlers.head()
```

```python
bowlers_most_wickets = bowlers.groupby('bowler', as_index=False).sum()[['bowler', 'wickets']]
bowlers_5_most_wickets = bowlers_most_wickets.nlargest(5, 'wickets')
bowlers_5_most_wickets
```

```python
def calculate_wicket_details(df) :
    bowler_list = []
    for row in df.itertuples() :
        matches = bowlers[bowlers.bowler == row.bowler]['bowler'].count()
        highest_wickets = bowlers[bowlers.bowler == row.bowler].max()[['match_id','wickets']]
        runs_in_match = dlv_new[(dlv_new.match_id == highest_wickets.match_id) & (dlv_new.bowler == row.bowler)].groupby('bowler', as_index=False).sum()[['total_runs']]
        bowler_list.append({'name' : row.bowler, 'wickets' : row.wickets, 'matches' : matches, 'best' : str(highest_wickets.wickets) + '/' + str(runs_in_match['total_runs'].iloc[0])})
        
    return bowler_list
```

```python
top5_bowler_details = pd.DataFrame(calculate_wicket_details(bowlers_5_most_wickets))
top5_bowler_details
```

```python
source2 = ColumnDataSource(top5_bowler_details)


plot2 = figure(x_range= top5_bowler_details.name, plot_height=450, title="bowlers with most wickets", toolbar_location=None)



plot2.vbar(x='name', top='wickets', width=0.6, source = source2, legend = 'name',
           fill_color=factor_cmap('name', palette=Spectral5, factors=top5_bowler_details.name))

plot2.xgrid.grid_line_color = None
plot2.y_range.start = 0
plot2.y_range.end = 180
hover = HoverTool(tooltips = [('Matches','@matches'),('Wickets','@wickets'), ('Best', '@best')])
plot2.add_tools(hover)
plot2.legend.orientation = "horizontal"
plot2.legend.location = "top_center"




show(plot2)
```

__Team Wins__

```python
team_runs_bymatch = dlv_new.groupby(['match_id','inning','batting_team','bowling_team'], as_index=False).sum()[['match_id', 'inning', 'batting_team', 'bowling_team', 'total_runs']]

team_runs_bymatch.head()
```

```python
team_win_bymatch = team_runs_bymatch.groupby('match_id').apply(lambda x: x.loc[x.total_runs.idxmax(),['inning', 'batting_team','total_runs']]).reset_index()
team_win_bymatch.head()
```

```python
team_wins_byinning= team_win_bymatch.groupby(['batting_team', 'inning'])['batting_team', 'inning'].agg({"wins": len}).reset_index()
team_wins_byinning
```

```python
team_list = team_wins_byinning['batting_team'].unique()
first_inning = team_wins_byinning[team_wins_byinning.inning == 1]['wins']
second_inning = team_wins_byinning[team_wins_byinning.inning == 2]['wins']
total_wins = np.add(first_inning, second_inning)

data = {'teams' : team_list,
        'first_bat'   : first_inning,
        'second_bat'   : second_inning,
        'total_wins' : total_wins
        }

source = ColumnDataSource(data=data)


p = figure(x_range=team_list, y_range=(0, 70), plot_height=550, title="Team wins by Batting Inning",
           toolbar_location=None, tools="")

p.vbar(x=dodge('teams', -0.25, range=p.x_range), top='first_bat', width=0.2, source=source,
       color="#c9d9d3", legend=value("1st Bat"))

p.vbar(x=dodge('teams',  0.0,  range=p.x_range), top='second_bat', width=0.2, source=source,
       color="#718dbf", legend=value("2nd Bat"))

hover = HoverTool(tooltips = [('Total Wins','@total_wins')])
p.add_tools(hover)


p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
